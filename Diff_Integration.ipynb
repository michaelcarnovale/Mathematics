```python
import numpy as np
import matplotlib.pyplot as plt
```

# Derivatives

Graphing the function $f(x) = x^3 + x^2 + x$ along with its first and second derivative.

Its first derivative is $\frac{dy}{dx} = 3x^2 + 2x + 1$.
Its second derivative is $\frac{d^2y}{dx^2} = 6x + 2$.

```python
from scipy.misc import derivative

def f(x):
    return x**3 + x**2 + 1*x

x = np.linspace(-2, 2, 1000)
y = f(x)
dydx = derivative(f, x, n = 1) # n sets the order of the derivative
dydx2 = derivative(f, x, n = 2)
plt.plot(x, y, label = 'f(x)')
plt.plot(x, dydx, label = 'dy/dx')
plt.plot(x, dydx2, label = 'd2y/dx2')
plt.legend()
```

```python
import sympy as sp

x = sp.symbols('x', real = True)
f = x**3 + x**2 + 1*x
dfdx = sp.diff(f, x, 1)
dfdx2 = sp.diff(f, x, 2)
print(dfdx) # First derivative
print(dfdx2) # Second derivative
```

# Integrals

Calculating the following definite integral 

$$\int_{-\infty}^{0} \frac{1}{\sqrt2\pi}e^\frac{-(x)^2}{2}dx$$

```python
def f(x):
    return np.exp((-(x)**2)/2) / np.sqrt(2*np.pi)

x = np.linspace(-3, 3, 1000)
y = f(x)
plt.plot(x, y, label = 'f(x)')
```

```python
from scipy.integrate import quad

integrand = lambda x: np.exp((-(x)**2)/2) / np.sqrt(2*np.pi)
integral = quad(integrand, -np.inf, 0)
print(integral)
```
